# Create a bar chart or histogram to visualize the distribution of a categorical or continous variable , such as the distribution of ages or genders in a population.

```python
import xml.etree.ElementTree as ET
import pandas as pd
import matplotlib.pyplot as plt

tree = ET.parse("datasets/API_SP.POP.TOTL_DS2_en_xml_v2_121612.xml")
root = tree.getroot()

data = []
for rec in root.findall(".//record"):
    c = rec.find("./field[@name='Country or Area']")
    y = rec.find("./field[@name='Year']")
    v = rec.find("./field[@name='Value']")
    if c is not None and y is not None and v is not None and v.text:
        data.append([c.text, int(y.text), float(v.text)])

df = pd.DataFrame(data, columns=["Country","Year","Population"])
```

## Bar Chart - Afghanistan

```python
df_country = df[df["Country"]=="Afghanistan"].sort_values("Year").head(10)

plt.figure()
plt.bar(df_country["Year"], df_country["Population"])
plt.title("Bar Chart - Afghanistan")
plt.xlabel("Year")
plt.ylabel("Population")
plt.xticks(rotation=45)
plt.show()
```
